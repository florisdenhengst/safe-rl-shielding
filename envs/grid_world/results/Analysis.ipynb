```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

UNSHIELDED = 'unshielded_reset'
EXACT = 'exact_reset'
EXACT_SHAPED = 'shaped_reset'

REWARDS_FILE = 'avg_reward.data'
EPLEN_FILE = 'episodelen.data'
```

```python
names=['len',]

unshielded = pd.read_csv(UNSHIELDED + '_' + EPLEN_FILE, names=names)
exact = pd.read_csv(EXACT + '_' + EPLEN_FILE, names=names)
```

```python
exact[:200].plot()
plt.show()
```

```python
unshielded[:200].plot()
plt.show()
```

```python
data = unshielded.merge(exact, how='outer', left_index=True, right_index=True, suffixes=('_u', '_s'))
data = data.rename(columns={'len_u': 'unsafe', 'len_s': 'shielded'})
data[:100].plot()
plt.xlabel('episode')
plt.ylabel('episode length')
plt.show()
```

```python
data[60:100].plot()
plt.xlabel('episode')
plt.ylabel('episode length')
plt.show()
```

```python
data[:100].plot(logy=True)
plt.xlabel('episode')
plt.ylabel('episode length (logscale)')
plt.show()
```
